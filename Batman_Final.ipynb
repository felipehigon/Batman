# VILLAINS

`````
#
#                   *********************
#                        <Bat>Villains
#                   *********************
#Comments: Team Machine's learning has thought that Joker and Joker Villain classes could be inherited classes from Villain class. 

class Villain:
  def __init__(self, we):
    self.ko=False
    if isinstance(we, bool):
      self.we=we
    else:
      raise TypeError('ERROR: we must be boolean.')
  def evaluation(self):
    print(f'This human has K.O state equal to {self.ko} and weapons state equal to {self.we}') 
  
class Joker(Villain):
  def evaluation(self):
    print(f'Joker K.O state: {self.ko}. Weapons possesion state: {self.we}')
    
class Joker_Villain(Villain):
  def evaluation(self):
    print(f'Human K.O state: {self.ko}. Weapons possesion state: {self.we}')


`````

# WEAPONS

`````
#Comments: This time for explain Batman's electromagnetic pulses and Batman's rockets have designed a main class
#called Weapon from which are inherited two classes; EM_Pulse and Rocket. Both  classes (EM_Pulse and Rocket) have  
#one more inherited class, whose names are Bat_Rocket and Bat_EM_Pulse. The classes we aimed to explain.
#
#
#                   *********************
#                        <Bat>Weapons
#                   *********************

#Parent class: Weapon
class Weapon:
  #Weapon Actions
  def __init__(self):
    self.dict_weapon_actions = {"load": self._load, "fire": self._fire}

  #Perform an action
  def op_weapon(self, weapon_action):
    if weapon_action == "load":
      self._load()
    elif weapon_action == "fire":
      self._fire()
    else:
      raise ValueError("ERROR: Unknown weapon action. Undefined or non-weapon action")  

  def _load(self):
    print('weapon-ammo loaded') 

  def _fire(self):
    print('weapon-ammo fired') 


#The Rocket and EMP (Electro Magnetic Pulse) are the ammunition of the Batmobile, 
#the car can only shoot the weapons if it has ammunition in the arsenal.

#Derived class 1: EM_Pulse
class EM_Pulse(Weapon):  
  def __init__(self):
    self.wtype = "Pulse" 

  #OVERRIDE (specific Weapons update)
  def _load(self):
    print('EM_Pulse loaded') 

  #OVERRIDE (specific Weapons update)
  def _fire(self):
    print('EM_Pulse fired') 
  
#Final class 1: Bat_EM_Pulse
class Bat_EM_Pulse(EM_Pulse):  
  def __init__(self):
    self.wtype = "Pulse" 
    self.wsubtype = "Bat_Weapon" 

  #OVERRIDE (specific Bat-Weapons® update)
  def _load(self):
    print('Bat_EM_Pulse® loaded') 

  #OVERRIDE (specific Bat-Weapons® update)
  def _fire(self):
    print('Bat_EM_Pulse® fired') 

#Derived class 2: Rocket
class Rocket(Weapon):
  def __init__(self):
    self.wtype = "Projectile" 

  #OVERRIDE (specific Weapons update)
  def _load(self):
    print('Rocket loaded') 

  #OVERRIDE (specific Weapons update)
  def _fire(self):
    print('Rocket fired') 
  
#Final class 2: Bat_Rocket
class Bat_Rocket(Rocket):  
  def __init__(self):
    self.wtype = "Projectile" 
    self.wsubtype = "Bat_Weapon" 

  #OVERRIDE (specific Bat-Weapons® update)
  def _load(self):
    print('Bat_Rocket® loaded') 

  #OVERRIDE (specific Bat-Weapons® update)
  def _fire(self):
    print('Bat_Rocket® fired') 
`````

# VEHICLES

`````
#
#                   *********************
#                       <Bat>Vehicles
#                   *********************
#Vehicles Parent class 1: Super_Car
#For generating Batmobile class two parent classes have been designed. 
#   1. Super_Car: Class with the car properties that Batmobile has.
#   2. Autonomous_Tank: Class with the armamentistic properties that Batmobile has.  
#The idea is apply multi-inheritance for generating Batmobile class with properties from both classes.

class Super_Car(object):
  def __init__(self, pilot, liters):
    #Init pilot
    if isinstance(pilot, str):
      if (pilot=="Batman") or (pilot=="Robin"):
        self.pilot = pilot
      else:
          raise ValueError("pilot must be a non-empty string")
    else:
      raise TypeError("pilot must be an str value")

    #Init liters
    if isinstance(liters, float) & (liters > 0.0):
      self.liters = liters
    else:
      raise TypeError("liters must be a float value")   

    #Drive Actions
    self.dict_drive_actions = {"drive": self._drive, "drive_turbo": self._drive_turbo}

  #Perform drive
  def run_driving(self, dist, drive_type="drive"):
    if drive_type == "drive":
      self._drive(dist)
    elif drive_type == "drive_turbo":
      self._drive_turbo(dist)
    else:
      raise ValueError("ERROR: Unknown drive type. Undefined or non-drive action")       

  # we assume consumption of 10 liters/100 km at 120 km/h
  def _drive(self, dist):
    if isinstance (dist, int):
      self.dist = dist
    else:
      raise Exception("dist must be int")
    if (self.dist/10) <= self.liters:
      self.liters = self.liters - (self.dist/10)
      print(f"gas left: {self.liters} liters")
    else:
      print(f"not enough gas for: {self.dist} km")

  # we assume consumption of 30 liters/100 km at 240 km/h
  def _drive_turbo(self, dist):
    if isinstance (dist, int):
      self.dist = dist
    else:
      raise Exception("dist must be int")
    if self.dist*0.3 <= self.liters:
      self.liters = self.liters - (self.dist*0.3)
      print(f"gas left: {self.liters} liters")
    else:
      print(f"not enough gas for: {self.dist} km")


#Vehicles Parent class 2: Autonomous_Tank
class Autonomous_Tank(object):
  
  def __init__(self, rockets, EMPs):
    #Init rockets
    self.rockets = []
    
    if isinstance(rockets, list):
      for rocket in rockets:
        if isinstance(rocket, Rocket):
          self.rockets.append(rocket)
          rocket.op_weapon('load')
        else:
          raise TypeError("rockets must contain all-type Rocket members")

    else:
      raise TypeError("rockets must be a list")

    #Init EMPs
    self.EMPs = []
    if isinstance (EMPs, list):
      for emp in EMPs:
        if isinstance(emp, EM_Pulse):
          self.EMPs.append(emp)
          emp.op_weapon('load')
        else:
          raise TypeError("EMPs must contain all-type EM_Pulse members")
    else:
      raise TypeError("rockets must be a list")

    #Shoot Actions
    self.dict_shoot_actions = {"shoot_rocket": self._shoot_rocket, "shoot_emp": self._shoot_emp}

  #Perform shoot
  def run_shooting(self, dist, villain, shoot_type="shoot_rocket"):
    if shoot_type == "shoot_rocket":
      self._shoot_rocket(villain)
    elif shoot_type == "shoot_emp":
      self._shoot_emp(villain)
    else:
      raise ValueError("ERROR: Unknown shoot type. Undefined or non-shoot action")   

  def _shoot_rocket(self, villain):
    if isinstance(villain, Villain):
      if len(self.rockets) == 0:
        print('NO AMMO!!!!')
      else:
        self.rockets.pop()
        villain.ko=True
        print(f"ROCKET FIRED. Number of rockets left: {len(self.rockets)}")    
    else:
      raise TypeError('ERROR: Villain must be Villain member.')

  def _shoot_emp(self, villain):
    if isinstance(villain, Villain):
      if len(self.EMPs) == 0:
        print('NO AMMO!!!!')
      else:
        self.EMPs.pop()
        villain.we=False
        print(f"EMP FIRED. Number of EMPs left: {len(self.EMPs)}")
    else:
      raise TypeError('ERROR: Villain must be Villain member.')

    


# Hybrid class: Batmobile® Armoured Vehicle Unit

class Batmobile(Super_Car, Autonomous_Tank):
  def __init__(self, bat_rockets, bat_EMPs, pilot="Batman", liters=150.0):

    #The actual Batmobile could also act as an unmanned vehicle, but given the problem  
    #especifications this version must be driven by Batman or Robin only. 

    if (pilot == "Batman") or (pilot == "Robin"):
      Super_Car.__init__(self, pilot, liters)
    else:
      raise ValueError("pilot only must be from two valid values: Batman or Robin")

    if isinstance(bat_rockets, list):
      self.bat_rockets=bat_rockets
      for bat_rocket in bat_rockets:
        if isinstance(bat_rocket, Bat_Rocket):
          pass
        else:
          raise TypeError("bat_rockets must contain all-type Bat_Rocket members")
      
      if isinstance(bat_EMPs, list):
        self.bat_EMPs=bat_EMPs
        for bat_EMP in bat_EMPs:
          if isinstance(bat_EMP, Bat_EM_Pulse):
            pass
          else:
            raise TypeError("bat_EMPs must contain all-type Bat_EM_Pulse members") 
        Autonomous_Tank.__init__(self, bat_rockets, bat_EMPs)   
      else:
        raise TypeError("bat_EMPs must be a list") 
    else:
      raise TypeError("bat_rockets must be a list")

  #OVERRIDE parent _drive (compsumption update for this armoured super car)
  # we assume a new consumption of 20 liters/100 km at 120 km/h

  def _drive(self, dist):
    if isinstance (dist, int):
      self.dist = dist
    else:
      raise Exception("dist must be int")
    if (self.dist/20) <= self.liters:
      self.liters = self.liters - (self.dist/20)
      print(f"gas left: {self.liters} liters")
    else:
      print(f"not enough gas for: {self.dist} km")
  


  #OVERRIDE parent _drive_turbo (compsumption update for this armoured super car)
  # we assume a new consumption of 40 liters/100 km at 240 km/h

  def _drive_turbo(self, dist):
    if isinstance (dist, int):
      self.dist = dist
    else:
      raise Exception("dist must be int")
    if self.dist*0.4 <= self.liters:
      self.liters = self.liters - (self.dist*0.4)
      print(f"gas left: {self.liters} liters")
    else:
      print(f"not enough gas for: {self.dist} km")

  #OVERRIDE parent _shoot_rocket (specific Bat-Weapons® update)

  def _shoot_rocket(self, villain):

    if isinstance(villain, Villain):
      if len(self.bat_rockets) == 0:
        print('NO BAT-ROCKETS-AMMO!!!!')
      else:
        self.bat_rockets[0].op_weapon("fire")
        self.bat_rockets.pop()
        
        villain.ko=True
        villain.evaluation()
        print(f"BAT-ROCKET FIRED. Number of Bat-Rockets® left: {len(self.bat_rockets)}")

    else:
      raise TypeError('ERROR: Villain must be Villain member.')
        

  def _shoot_emp(self, villain):
    if isinstance(villain, Villain):
      if len(self.bat_EMPs) == 0:
        print('NO EMP-AMMO!!!!')
      else:
        self.bat_EMPs[0].op_weapon("fire")
        self.bat_EMPs.pop()
        villain.we=False
        villain.evaluation()
        print(f"BAT-EMP FIRED. Number of Bat-EMPs® left: {len(self.bat_EMPs)}")
    else:
      raise TypeError('ERROR: Villain must be Villain member.')
`````
